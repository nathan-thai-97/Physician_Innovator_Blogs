### Workaround for JupyterLab Terminal

When using Jupyter Lab, **the terminal will be occupied**. Here are a few things I can do to manage that:

- Use a Terminal Multiplexer (tmux): Tmux allows you to have multiple "windows" or "panes" inside a single terminal tab.

    1. Install it: ```sudo apt install tmux``` (Linux).

    3. Start a session: Type ```tmux```.

    4. Run Jupyter: Inside tmux, type ```jupyter lab```.

    5. "Detach" from the session: Press ```Ctrl + B```, then let go and hit ```D```.

        Your terminal is now completely free! Jupyter is running in a hidden layer. To go back and check on it: Type tmux attach.

- Built-in Terminal in Jupyter Lab
    1. Run jupyter lab normally.

    2. In the Launcher (the "plus" icon tab), look for the Terminal icon.

    3. This opens a fully functional terminal window right next to your notebooks.

        This is great for staying organized because the Jupyter terminal automatically starts in your project's root directory, so you don't have to cd everywhere.

### Git Pull Trick for Jupyter Lab

**Option 1**: "I want to keep my changes and the remote changes" (Recommended)
Use a stash. This essentially puts your work in a temporary "closet," pulls the new code, and then brings your work back out to sit on top of it.

```Bash
git stash
git pull
git stash pop
```
Note: If there is a "conflict" after the last step, you'll have to open the notebook and manually decide which cells to keep.

**Option 2**: "I don't care about my local changes; just give me the server version"
If you were just experimenting and want to throw away your local edits to Bear_Classifier.ipynb and match the repository exactly:

```Bash
git checkout -- models/Bear_Classifier.ipynb
git pull
```

**Option 3**: "I want to save my work properly before pulling"
If your changes are important, commit them first. This creates a permanent point in your history before the merge happens.

```Bash
git add models/Bear_Classifier.ipynb
git commit -m "Save local updates to bear classifier"
git pull
```
If you use this, Git will likely open a text editor asking you to write a "Merge message." Just save and exit that screen to finish.

### Images Data Options
I am trying to follow both Chapter 2 JNB and lesson 2 video but it has been a pain trying to collect images.

First, the book recommended using Bing (Azure API), then the video changed to DuckDuckGo, but the code keep throwing more and more error due to anti scrapping measures. Microsoft retired Bing Search API in 2025 as well 

Here are the best alternatives for gathering image data for your deep learning model:

1. ***ICrawler*** (The "Swiss Army Knife")
    icrawler is a powerful framework that can crawl Google, Bing, and Baidu all at once. It’s much more robust than the simple fastbook wrapper because it has built-in retry logic.

    How to use it:

    ```Python
    %pip install icrawler
    from icrawler.builtin import BingImageCrawler

    # This will download 50 images of 'banh mi' directly into a folder
    google_crawler = BingImageCrawler(storage={'root_dir': 'data/banh_mi'})
    google_crawler.crawl(keyword='banh mi', max_num=50)
    ```
    
    Pros: Very stable; multiple engines.


    Cons: A bit slower than a direct API.

3. ***Pre-made Datasets*** (The "Gold Standard")
    Instead of scraping the messy web, you can use datasets where the "Quality Control" has already been done for you. For a calorie calculator, these are perfect:
    
    Food-101: 101,000 images of 101 food categories.
    
    How to get it in fastai: ```path = untar_data(URLs.FOOD)```
    
    ***Open Images*** (by Google): Millions of images with bounding boxes (great if you want to detect the size of the steak, not just that it's a steak).
    
    ***Kaggle***: Search for "Nutrition" or "Food" datasets. There are many specifically for Vietnamese cuisine or Western fast food.

4. ***Local Scraping*** with "Simple Image Download"
    There is a lightweight library called simple_image_download that uses a different scraping logic than ddgs.
    
    ```Python
    %pip install simple_image_download
    from simple_image_download import simple_image_download as simp
    
    response = simp.simple_image_download
    response().download('grizzly bear', 10)
    ```
