```python
from ddgs import DDGS
from fastcore.all import L

def search_images(term, max_images=30):
    print(f"Searching for '{term}'...")
    with DDGS() as ddgs:
        # 'query' is the required first argument now
        results = ddgs.images(term, max_results=max_images)
        return L(results).itemgot('image')

# Try it now
ims = search_images('grizzly bear', max_images=1)
if len(ims) > 0:
    print(f"Success! Found {len(ims)} image(s)")
    print(f"First URL: {ims[0]}")
```

```python
dest = 'images/grizzly.jpg'
download_url(ims[0], dest, show_progress=FALSE)
```
